```
#     1. [25] Write Python code to perform the following tasks. (1.A, 1.B, 1.C, 1.D : below )
import pandas as pd
import io

#     (1.A)Read data from hwk06-01.csv into a DataFrame df1.
df = pd.read_csv(io.BytesIO(uploaded['hw06-01.csv']), header = None, names = ('A','B','C','D','E','F'))
```

```
#     (1.B)Create a new DataFrame df2 by replicating each row of df1 with the number of copies
#     as indicated in the count column. For example, the first row in df1 should appear 30
#     times in df2, second row in df1 40 times in df2, etc.
#df.F.apply(str)
df["F"] = pd.to_numeric(df["F"],errors='coerce')
df2 = pd.DataFrame(columns = ['Department' , 'Status', 'Age' , 'Salary'])
rowCount = 0
#df2.loc[1] = ['Sales', 'Senior', '31..35', '46k..50k', '30']

for index, row in df.iterrows():
  #skip first row
  if index > 0:
    count = int(row['F'])
    count = int(count)
    for x in range(count):
#     (1.D) Make sure that df2 does not have the column count.
      df2.loc[rowCount] = [row['B'], row['C'], row['D'], row['E']]
      rowCount = rowCount + 1

#df2
#Check data types
#df.dtypes

#     (1.C)Perform the one-of-kinds encoding on the categorical columns. For example, if a categorical attribute has three values “A”, “B”, and “C”, you will encode “A” as 0, “B” as
#     1, and “C” as 2. One way to do this in Python is to save the values in a list and use the
#     list index as the coding.
mapping_dict = {
    "Department": {
        "Sales": 0,
        "Systems": 1,
        "Marketing": 2,
        "Secretary": 3
    },
    "Status": {
        "Senior": 0,
        "Junior": 1
    },
    "Age": {
        "21..25": 0,
        "26..30": 1,
        "31..35": 2,
        "36..40": 3,
        "41..45": 4,
        "46..50": 5
    },
    "Salary": {
        "26k..30k": 0,
        "31k..35k": 1,
        "36k..40k": 2,
        "41k..45k": 3,
        "46k..50k": 4,
        "66k..70k": 5
    }
}
df2 = df2.replace(mapping_dict)
df2
#Checking CSV files
#df2.to_csv("df2.csv")
#files.download('df2.csv')
```

```
#     2. [25] Write Python code to use sklearn.tree.DecisionTreeClassifier to learn a decision tree
#     using the df2 as the training data, and then use the decision tree to predict the status of a user
#     provided unseen data, such as:
#                    t =< department : systems, status :?, age : 28, salary : 50K >

from sklearn.tree import DecisionTreeClassifier
#from sklearn.model_selection import train_test_split

#split dataset in features and target variable
feature_cols = ['Department' , 'Age' , 'Salary']
X = df2[feature_cols]
y = df2['Status'].values


#Split dataset into training set and test set
#X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

#Create Decision Tree classifer object
clf = DecisionTreeClassifier()

#Train Decision Tree Classifer
clf = clf.fit(X, y)

#Create t
t = [[1,1,4]]

#Predict the response for test dataset
clf.predict(t)
```

```
#     4. [25] Write Python code to use sklearn.naive_bayes to learn a Guassian Naive Bayes classifier
#     using df3 as the training data, and use the learned predictive model to predict the status of a
#     user provided unseen data, for example,
#                    t =< department : systems, status :?, age : 28, salary : 50K >

from sklearn.naive_bayes import GaussianNB

#Instantiate the classifier
gnb = GaussianNB()

#split dataset in features and target variable
feature_cols = ['Department' , 'Age' , 'Salary']
X = df3[feature_cols]
y = df3['Status']

#Train Guassian Naive Bayes classifier
gnb = gnb.fit(X, y)

#Create t
t = [[1,28,50000]]
#Predict the response for test dataset
gnb.predict(t)
```
